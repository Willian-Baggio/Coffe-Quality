**Entendendo o Dataset**

Feito a remoção de colunas, agora sobraram apenas 15 colunas que serão utilizadas para realizar a análise da qualidade do café.

* Country of Origin: Refere-se ao país de origem do café.

* Processing Method: Método de processamento do café.

* Aroma: Refere-se ao cheiro ou fragrância do café.

* Flavor: O café é avaliado com base no sabor, incluindo doçura, amargor, acidez e outras notas de sabor.

* Aftertaste: Refere-se ao sabor persistente que permanece na boca após engolir o café.

* Acidity: A acidez no café refere-se ao brilho ou vivacidade do sabor.

* Body: O corpo do café refere-se à espessura ou viscosidade do café na boca.

* Balance: Equilíbrio refere-se a quão bem os diferentes componentes do sabor do café funcionam juntos.

* Overall: Qualidade geral do café.

* Total Cup Points: Total de 10 características.

* Moisture Percentage: Porcentagem de umidade.

* Sweetness: Pode ser descrito como semelhante a caramelo, frutado ou floral, e é uma qualidade desejável no café.

* Category Defects:
  * Category One Defects: Defeitos primários que podem ser percebidos através da inspeção visual dos grãos de café.

  * Category Two Defects: Defeitos secundários, só podem ser detectados por meio da degustação. Esses defeitos incluem fermentação excessiva, ranço, sabor químico, etc.

# **Analisando os dados**

```python
paises_contagem = df['Country of Origin'].value_counts()

sorted_countries = paises_contagem.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=sorted_countries.index, y=paises_contagem, palette='gray')

plt.xlabel('País de Origem')
plt.ylabel('Contagem de Amostras')
plt.title('Amostras de café por País')
plt.xticks(rotation=90)
plt.show()
```

```python
agrupamento = df.groupby('Country of Origin')['Total Cup Points'].mean()

sorted = agrupamento.sort_values(ascending=False)

cols = ['grey' if (x < max(agrupamento)) else 'orange' for x in agrupamento]

plt.figure(figsize=(12, 6))
sns.barplot(x=agrupamento.index, y=agrupamento.values, palette=cols)

plt.xlabel('País de Origem')
plt.ylabel('Média Total Cup Points')
plt.title('Média Total Cup Points por País')
plt.xticks(rotation=90)
plt.show()
```

```python
cafe = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 'Overall']

df[cafe].hist(color="skyblue", bins=20, figsize=(6, 6), layout=(3, 3))

plt.suptitle("Histogramas das Características do Café", fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
defect_columns = ['Category One Defects', 'Quakers', 'Category Two Defects']

plt.figure(figsize=(12, 6))
for col in defect_columns:
    defect_group = df.groupby(col)[defect_columns].mean()
    defect_group.plot(kind='bar', colormap='viridis', alpha=0.8, ax=plt.gca())
plt.xlabel('Presença de Defeitos')
plt.ylabel('Média das Características')
plt.title('Média das Características em relação à Presença de Defeitos')
plt.legend(title='Defeitos', loc='upper left', labels=['Category One Defects', 'Quakers', 'Category Two Defects'])
plt.xticks(rotation=0)
plt.show()
```

```python
fig, axs = plt.subplots(len(cafe), 1, figsize=(12, 20), sharex=True)

for i, coluna in enumerate(cafe):
    sns.boxplot(x=df['Processing Method'], y=coluna, data=df, palette='pastel', width=0.5, ax=axs[i])
    axs[i].set_ylabel(coluna)
    axs[i].set_title(f'Boxplot de {coluna} por Método de Processamento')
    axs[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
agrupamento = df.groupby('Color')['Moisture Percentage'].mean()

min_index = agrupamento.idxmin()

cols = ['red' if x == min_index else 'grey' for x in agrupamento.index]

plt.figure(figsize=(12, 6))
sns.barplot(x=agrupamento.index, y=agrupamento.values, palette=cols)

plt.xlabel('Cor do Café')
plt.ylabel('Média de Umidade')
plt.title('Média de Umidade por Cor de Café')
plt.xticks(rotation=45)
plt.show()
```
